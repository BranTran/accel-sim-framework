```python
list_parquet = glob.glob(os.path.join(os.getcwd(),"..","..",".mch","**","*3035*","**","*.par*"), recursive=True)
```

```python
pd.read_parquet(list_parquet[0]).head()
```

```python
result=pd.DataFrame()

for file in list_parquet:
    df = pd.read_parquet(file)
    # print(df)
    a=df.pivot_table(index=["timestamp","hostname","device"],columns='sensor_name', values='int64')
    # print(a)
    b=df.pivot_table(index=["timestamp","hostname","device"],columns='sensor_name', values='float64')
    temp=pd.merge(a, b, how='inner', on=["timestamp","hostname","device"])
    result=pd.concat([result,temp])
result
```

```python
statics_str = '''-                 <param name="static_cat1_flane" value="15.29035866"/>
-                 <param name="static_cat1_addlane" value="0.586233603"/> 
-                 <param name="static_cat2_flane" value="18.6179906"/>
-                 <param name="static_cat2_addlane" value="0.645228013"/>
-                 <param name="static_cat3_flane" value="19.10017723"/>
-                 <param name="static_cat3_addlane" value="0.726863055"/> 
-                 <param name="static_cat4_flane" value="18.55029744"/>
-                 <param name="static_cat4_addlane" value="0.6099397"/>
-                 <param name="static_cat5_flane" value="14.74826681"/>
-                 <param name="static_cat5_addlane" value="0.514367937"/>
-                 <param name="static_cat6_flane" value="48.94875596"/>
-                 <param name="static_cat6_addlane" value="0.0"/> 
'''
# -                 <param name="static_light_flane" value="1.965373811"/> 
# -                 <param name="static_light_addlane" value="0.003966868"/>
# -                 <param name="static_intadd_flane" value="19.70468506"/> 
# -                 <param name="static_intadd_addlane" value="0.388578623"/>
# -                 <param name="static_intmul_flane" value="16.64811823"/> 
# -                 <param name="static_intmul_addlane" value="0.281803166"/>
statics_list = statics_str.splitlines()
import re
# Parse through the list of strings
flane_list = []
addlane_list = []
for string in statics_list:
    float_value = re.findall(r'value="(\d+\.\d+)"', string)
    if float_value:
        float_value = float(float_value[0])
        if "flane" in string:
            flane_list.append(float_value)
        if "addlane" in string:
            addlane_list.append(float_value)
flane_list,addlane_list
```

```python
import statistics
print(statistics.geometric_mean(flane_list))
print(statistics.geometric_mean([value for value in addlane_list if value != 0]))
```
